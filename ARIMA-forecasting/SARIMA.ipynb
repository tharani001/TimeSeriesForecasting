### SARIMA stands for Seasonal Autoregressive Integrated Moving Average Model. `It takes into account the seasonality in time series`

### `SARIMA has 7 Parameters` - (p,d,q) = 3 regular ARIMA components and (P,D,Q,m) =  4 seasonal components


P , D , Q are the seasonal order of autoregressive portion, integrated portion and Moving average portion.

`m denotes the frequency: the number of timesteps per seasonal cycle`
