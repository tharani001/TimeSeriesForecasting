### VAR, VARMA, VARMAX are the multivariate models used to forecast multiple endogeneous variables especially when `there is two-way relationship` between them.

### If no two-way relationship exists, it is better to use univariate(ARIMA) models.

### `VAR is just the generalized version of ARIMA to allow multiple time series.`

### `VAR(2) Model:`



### It assumes that `each series has an impact on one another`, we need to verify if it is true using `Granger-Casuality Test.`

### The Granger Causality Test checks whether the past values of one time series (T1) help predict another time series (T2) and vice versa. It determines if there is a causal relationship between two time series.
